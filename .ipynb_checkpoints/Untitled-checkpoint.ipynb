```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as s

df = pd.read_csv('PandasPractice/pokemon_data.csv')
s.set_style('whitegrid')
```

```python
s.distplot(df['Speed']) #distribution of the data
# y axis has brobablity
```

```python
s.distplot(df['Speed'], hist=False)
mp.show()
```

```python
# two columns: Bivariate Distributions
s.jointplot('Speed', 'Attack', df)
mp.show
```

```python
df = df[(df.Speed < 125 ) & ( df.Attack <150 )]
s.jointplot('Speed', 'Attack',df)
mp.show()
```
