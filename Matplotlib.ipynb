```python
x = np.linspace(1, 10, 100)

mp.figure(1)

mp.subplot(3, 4, 1)
mp.title('Squared')
mp.plot(x, x**2)

mp.subplot(3, 4, 2)
mp.title('Cubed')
mp.plot(x, x**3)

mp.subplot(3, 4, 3)
mp.title('Quad')
mp.plot(x, x**4)

mp.figure(2)

mp.subplot(3, 4, 1)
mp.title('Five')
mp.plot(x, x**5)

mp.subplot(3, 4, 2)
mp.title('Six')
mp.plot(x, x**6)

mp.subplot(3, 4, 3)
mp.title('Seven')
mp.plot(x, x**7)

mp.figure(3)

mp.subplot(3, 4, 1)
mp.title('Eight')
mp.plot(x, x**8)

mp.subplot(3, 4, 2)
mp.title('Nine')
mp.plot(x, x**9)

mp.subplot(3, 4, 3)
mp.title('Ten')
mp.plot(x, x**10)
```
